```python
from pylab import scatter
import pylab
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import LogFormatter
from matplotlib.ticker import FuncFormatter
import math
```

```python
df = pd.read_csv('papers_summary_04-18-2014.csv',sep = '\t')
df.head()
```

```python
# rename the columns
df.rename(columns = {df.columns[0]:'Citation_Key',df.columns[1]:'Citation_Key_Plus',df.columns[3]:'frequency_(Hz)', df.columns[4]: 'Se_(V/m)^2/Hz',
                     df.columns[5]:'wSe_(V/m)^2',df.columns[6]:'w^1.5Se_(V/m)^2.sqrt(Hz)', df.columns[7]:'d_(um)',
                     df.columns[8]:'T_(K)', df.columns[9]:'trap_type', df.columns[10]:'Plot',df.columns[10]:'Reference_No',
                     df.columns[11]:'Full_Reference'},inplace = True)
df.head()
```

```python
# create one additional column for Reference_No, with a number instead of string 
# this is used for sorting by reference in the figure label
reflst = []
for ind in df.index:
    if isinstance(df['Reference_No'][ind],float):# == u'NA':
        reflst.append(np.NaN)
    elif df['Reference_No'][ind] == u'this work':
        reflst.append(1000)
    else:
        #print type (df['Reference_No'][ind])
        reflst.append(int(df['Reference_No'][ind]))
sr0 = pd.Series(reflst, name = 'Numeric_Reference_No')
df = df.join(sr0)
#df.head(10)  
```

```python
markerdict = {'triangle_down': 'v','triangle_up': '^', 'triangle_left': '<', 'circle': 'o', 'triangle_right': '>',	
              'square': 's',  	
            'pentagon': 'p',  'star': '*', 'plus': '+',  'x': 'x',  	
            'diamond': 'D', 'thin_diamond': 'd'} 
#'octagon':'8', 'tri_down':'1', 'tri_up':'2', 'tri_left':'3', 'tri_right':'4', 'vline':'|', 'hline': '_', 'hexagon1':'h', 'hexagon2':'H',

referencedict = {'Allcock2010':'pentagon', 'Allcock2013':'thin_diamond', 'Amini2010':'triangle_right', 'Brown2011':'thin_diamond',
                 'Chiaverini2014':'circle', 'Daniilidis2010':'triangle_up', 'Daniilidis2013':'square', 'Doret2012':'triangle_left',
                 'Epstein2007':'star', 'Hite2012':'triangle_down', 'Labasciewicz2008a':'x', 'Leibrandt2009':'plus', 'Leibrandt2009_cold':'star',
                 'Seidelin2006':'diamond', 'Wang2010':'circle', 'Talukdar2014':'pentagon', 'Niedermayr2014':'triangle_right', 'default':'plus'}
```

```python
# create two additional columns in the dataframe
#
# the first column denotes the marker to be used in the plot
# use one marker per paper
# the markers for papers which are not plotted do not matter
# if a citation has different temperatures, use the same marker for all tempreatures
#
# the second column denotes the color to be used in the plot
# for temperature > 200 K use black, and if you run into a repeated marker use grey
# for temperature < 200 K use blue, and if you run into a repeated marker use green

citation = df.Citation_Key[0] # assume that the first entry in the table is plotted
mrklist  = []
clrlist = []
mrkindex_h = 0
mrkindex_c = 0
shft = 0
for ind in df.index:
    if df.Plot[ind] and citation != df.Citation_Key[ind]:
        citation = df.Citation_Key[ind]
        if df['T_(K)'][ind] > 200:
            mrkindex_h += 1
        else:
            mrkindex_c += 1
    color_hot = 'black'
    color_cold = 'blue'
    if mrkindex_h > len(markerdict)-2:
        #shft = len(marker_dict)
        color_hot = 'grey'
    if mrkindex_c > len(markerdict)-2:
        color_cold = 'cyan'
    try:
        whichmarker = referencedict[df.Citation_Key[ind]]
    except:
        whichmarker = referencedict['default']
    #mrklist.append(markerdict[markerdict.keys()[mrkindex-shft]])
    mrklist.append(markerdict[whichmarker])
    if df['T_(K)'][ind] > 200:
        clrlist.append(color_hot)
    else:
        clrlist.append(color_cold)

sr1 = pd.Series(mrklist, name = 'Marker')
sr2 = pd.Series(clrlist, name = 'Color')
print sr1.shape
print sr2.shape
print df.shape
df2 = df.join(sr1)
df2 = df2.join(sr2)
print df2.shape
#df2.head()
```
